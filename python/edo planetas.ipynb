```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler(r0,v0,t0=0,dt=0.001,nsteps=1000, G=4*np.pi**2):
    x, y, z = [r0[0]], [r0[1]], [r0[2]]
    x0 = r0[0]
    y0 = r0[1]
    z0 = r0[2]
    vx0 = v0[0]
    vy0 = v0[1]
    vz0 = v0[2]
    t = [t0]
    for i in range(1, nsteps):  
        #xi, yi, zi = r0 + v0*dt
        
        xi=x0+vx0*dt
        yi=x0+vy0*dt
        zi=z0+vz0*dt
        rp = np.sqrt(xi**2 + yi**2 + zi**2)
        vxi=vx0-G*x0*dt/(rp**3)
        vyi=vy0-G*y0*dt/(rp**3)
        vzi=vz0-G*z0*dt/(rp**3)
        
        
        #vxi, vyi, vzi = v0 - G*0.001*r0*dt/(rp**3)
        x.append(xi)
        y.append(yi)
        #z.append(zi)
        t.append(t0+dt*i)
        x0 = xi
        vx0 = vxi
        
        y0 = yi
        vy0 = vyi
        
        z0 = zi
        vz0 = vzi
    return x, y, t
```

```python
r0 = np.array([0.0508011, -0.4011642, -0.2195576])
v0 = np.array([0.022318, 0.004855, 0.000279])*365

x, y, t = euler(r0, v0)
```

```python
ax = plt.axes()

ax.plot(t, y)
```
